```python
import numpy as np

#Intialize X variables, weights and y
x = np.array([3,6,4])
w = np.random.randn(3)
y = 3

#Set learning rate and epochs
lr = 0.01
epochs = 100

#Intitialize an empty list to hold the losses
losses = []

for i in range(epochs):
    #Calculate z
    z = x.dot(w)

    #Apply the activation function
    a = 1 / (1 + np.exp(-z))

    #Calculate loss and and to list for plotting
    loss = 1/2 * (y-a)**2
    losses.append(loss)

    #Compute Gradients
    dl_da = -(y-a) #derivative of loss with respect to a
    da_dz = a * (1-a) #derivative of a with respect to z
    dz_dw1 = x[0] #derivative of z with respect to w1
    dz_dw2 = x[1] #derivative of z with respect to w2
    dz_dw3 = x[2] #derivative of z with respect to w3

    dl_dw1 = da_dz * dl_da * dz_dw1 #derivative off loss with respect to w1
    dl_dw2 = da_dz * dl_da * dz_dw2 #derivative off loss with respect to w2
    dl_dw3 = da_dz * dl_da * dz_dw3 #derivative off loss with respect to w3

    #Update Weights
    w -= lr * np.array([dl_dw1, dl_dw2, dl_dw3])

#Plot Function
import matplotlib.pyplot as plt
plt.plot(losses)
plt.show()
```
